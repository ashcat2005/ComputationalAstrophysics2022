```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import HTML
%matplotlib inline


#Aqui se define la funcin que se va a graficar
def f(x,x0):
    sigma2 = 20.
    return np.exp(-(x-x0)**2/(2*sigma2))

# Primero se crea el escenario en el que se realiza la animacin.
# En este caso una figura y unos ejes.
# Adems se escriben las etiquetas de los ejes ya que estas no cambiaran
# a lo largo de la animacin
fig, ax=plt.subplots()
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$f(t,x)$')

# En esta lista se almacenan los frames de la animacin
imgs = []

#Aqui esta el loop principal.
# En cada paso se crea una linea con la funcin localizada en un valor de x0
# y se escribe el titulo con el tiempodel frame
for i in np.linspace(0,50,100):
    line, = ax.plot(xrange,f(xrange, 30+i), color='crimson')#ax.plot(a[:,i])
    title = ax.text(2.,0.95,f'Time = {i:.2f}')
    # Aqui se agrega la imagen a la lista
    imgs.append([line, title])

# La funcin ArtistAnimation() une las imagenes de la lista para crear el
# objeto con la informacin de la animacin
anim = animation.ArtistAnimation(fig, imgs, interval=50, blit=False)

# Para visualizar la animacin en jupyter se debe transformar a otro formato.
# Por ejmplo qui se transforma a un video de javascript
HTML(anim.to_jshtml())
```
